# **Accessing Open source LLMs for Text Generation, Summarization and Chat Completion**

### **What are Large Language Models (LLMs)?**

LLMs (Large Language Models) are a type of artificial intelligence model designed to understand and generate human-like text. They are trained on massive datasets consisting of text from books, websites, articles, code, and other sources.

> **Analogy**: *Think of an LLM as a super-smart autocomplete that doesn‚Äôt just finish your sentences, but can answer questions, write stories, code, and hold a conversation ‚Äî like having an AI-powered assistant always ready!*

**Key Features of LLMs:**

- **Large** refers to the number of parameters (often billions or even trillions) used in the model.

- **Language** refers to natural human language like English, Hindi, or even programming languages like Python.

- **Model** means a mathematical system trained to predict and generate text.

**What LLMs Can Do:**

- Answer questions

- Translate languages

- Write essays, summaries, or poems

- Generate code

- Perform sentiment analysis

- Extract information from documents

- Power chatbots and virtual assistants

## **Some of the LLM Uses We Perform in This Lab**

1. **Text Generation:**
LLMs can create new text based on a prompt you provide.

 Example: If you ask the model, ‚ÄúWrite a short story about space,‚Äù it will generate a story based on that prompt.

2. **Summarization:**
LLMs can shorten long texts while keeping the main points.

  There are two types:

 **Extractive Summarization:** Picks important sentences directly from the text.

 **Abstractive Summarization:** Paraphrases and rewords the content to make it shorter.

 Example: It can summarize long articles or reports into short, easy-to-read versions.

3. **Chat Completion:**
 LLMs can carry on conversations, answering questions or chatting based on the context.

 Example: If you ask, ‚ÄúWhat‚Äôs the weather like today?‚Äù it can generate a relevant response based on the question.

### **A. Using Hugging Face Transformers**

This example shows how to run an open-source large language model (LLM) called Falcon-7B locally on your own machine using Hugging Face's Transformers library and PyTorch. The goal is to generate text based on an input prompt.

1. Install Dependencies

2. Text Generation with falcon-7b-instruct via Hugging Face

````python
# Import necessary libraries from the 'transformers' library
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

# Load tokenizer and model (specifying the open-access Falcon 7B instruction model for this demo)
model_name = "tiiuae/falcon-7b-instruct"  # The model used for text generation tasks

# Load the tokenizer for the model; tokenizers convert input text into token IDs
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load the pre-trained model; it will be used for causal language modeling (text generation)
# The 'torch_dtype' is set to 'auto' to automatically choose the appropriate precision for the model
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype="auto")

# Create a text generation pipeline using the loaded model and tokenizer
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Define a prompt for text generation
prompt = "Explain the difference between AI and Machine Learning."

# Generate text based on the input prompt, limiting the response to a maximum of 100 new tokens
result = pipe(prompt, max_new_tokens=100)

# Print the generated text, which will be the model's response to the prompt
print(result[0]["generated_text"])
````

 *Expected Response:*

 The model will generate a clear and concise explanation, stating that AI is a broad field where machines perform tasks requiring human-like intelligence, while Machine Learning (ML) is a subset of AI focused on using data and algorithms to improve performance over time.

*Code Explanation*

- ```AutoTokenizer.from_pretrained(model_name):``` Loads the tokenizer associated with the chosen pre-trained model, which is responsible for converting text into tokens (numeric representations).

- ```AutoModelForCausalLM.from_pretrained(model_name, torch_dtype="auto"):``` Loads the model itself. CausalLM stands for Causal Language Modeling, which is used for tasks like text generation. The torch_dtype="auto" argument ensures that the model uses an appropriate data type for better performance.

- ```pipeline("text-generation", model=model, tokenizer=tokenizer):``` Creates a pipeline for text generation. A pipeline is an easy-to-use wrapper that simplifies the process of applying a model to a task like text generation.

- ```pipe(prompt, max_new_tokens=100):``` Generates text based on the given prompt. The max_new_tokens=100 limits the generated text to 100 tokens beyond the original prompt.

- ```print(result[0]["generated_text"]):``` Prints the model's generated response. The result is a list, and we extract the generated text from the first item (result[0]) in the list.
